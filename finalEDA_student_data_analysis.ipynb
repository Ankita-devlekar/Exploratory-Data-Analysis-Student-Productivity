**2. Sleep Group vs Average Focus Score**

```python
# Create sleep groups
df['sleep_group'] = pd.cut(
    df['sleep_hours'],
    bins=[0,5,7,9,12],
    labels=['Poor (0-5)','Average (5-7)','Good (7-9)','Excellent (9+)']
)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x='sleep_group',
    y='focus_score',
    hue='sleep_group',
    data=df,
    order=['Poor (0-5)','Average (5-7)','Good (7-9)'],
    palette='viridis',
    legend=False
)

plt.title("Average Focus Score by Sleep Duration", fontsize=14, fontweight='bold')
plt.xlabel("Sleep Duration (Hours)")
plt.ylabel("Average Focus Score")

plt.grid(axis='y', linestyle='--', alpha=0.5)

# Label bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.tight_layout()
plt.show()
```

1. Focus increases with better sleep.

Students with **poor sleep (0–5 hrs)** have an average focus score of about **8.8**, while those with **good sleep (7–9 hrs)** have a higher average focus score of around **9.6**.

2. Moderate sleep also improves focus.

Students sleeping **5–7 hours** show an average focus score of approximately **9.4**, which is **higher than the poor sleep group.**

**3. Stress Level vs Productivity**

```python
# Create stress groups
df['stress_group'] = pd.cut(
    df['stress_level'],
    bins=[0,3,6,10],
    labels=['Low','Moderate','High']
)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x='stress_group',
    y='productivity_score',
    hue='stress_group',   # avoids seaborn warning
    data=df,
    order=['Low','Moderate','High'],
    palette='plasma',
    legend=False
)

plt.title("Average Productivity by Stress Level", fontsize=14, fontweight='bold')
plt.xlabel("Stress Level", fontsize=11)
plt.ylabel("Average Productivity Score", fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.5)

# Label bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=10)

plt.tight_layout()
plt.show()
```

1. Productivity decreases as stress increases.

Students with low stress show an average productivity score of **approximately ~78–80**, while students with high stress show a much lower productivity score of around **~55–60**.

2. Moderate stress shows intermediate productivity.

Students with moderate stress have an average productivity score of about **~65–70**, **indicating a gradual decline in productivity as stress levels increase.**

# STEP 10: Managerial Insights from Student Productivity Analysis(Q and A)

**1. Ideal Student Profile: Comparison of Top vs Bottom Performers**




Research Question:
What habits differentiate high-productivity students from low-productivity students?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create top and bottom groups
top_students = df[df['productivity_score'] >= df['productivity_score'].quantile(0.8)]
low_students = df[df['productivity_score'] <= df['productivity_score'].quantile(0.2)]

# Calculate mean values
top_mean = top_students.mean(numeric_only=True)
low_mean = low_students.mean(numeric_only=True)

compare_df = pd.DataFrame({
    'Top 20%': top_mean,
    'Bottom 20%': low_mean
}).T

# Select important columns
cols = ['study_hours_per_day','sleep_hours','total_screen_time','stress_level','focus_score']
compare_df = compare_df[cols]

compare_df.T.plot(kind='bar', figsize=(10,6))
plt.title("Top vs Bottom Students Profile")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
Top-performing students tend to have higher study hours, better sleep, lower screen time, lower stress levels, and higher focus scores compared to low-performing students.

**Managerial Implication:**
Encourage students to maintain balanced study hours, prioritize adequate sleep, reduce screen time, and manage stress to improve focus and overall productivity.

**2. Optimal Study Hours for Maximum Productivity**

Research Question:
What is the ideal range of study hours that maximizes student productivity?

```python
# Create bins
df['study_bins'] = pd.cut(df['study_hours_per_day'], bins=[0,2,4,6,8,12],
                         labels=['0-2','2-4','4-6','6-8','8+'])

# Plot
sns.barplot(x='study_bins', y='productivity_score', data=df)
plt.title("Optimal Study Hours vs Productivity")
plt.xlabel("Study Hours Range")
plt.ylabel("Average Productivity Score")
plt.show()
```

**Insight:**
Productivity increases with study hours and peaks in the 6–8 hours range, indicating this as the optimal study duration.

**Managerial Implication:**
Encourage students to study around 6–8 hours per day, as excessive or insufficient study time may lead to lower productivity.

**3. Impact of Different Screen Activities on Productivity**

Research Question:
Which type of screen activity impacts student productivity the most?

```python
activities = ['social_media_hours', 'youtube_hours', 'gaming_hours']
df[activities + ['productivity_score']].corr()['productivity_score'].drop('productivity_score').plot(kind='bar')
plt.title("Screen Activity Impact on Productivity")
plt.show()
```

**4. Impact of Attendance and Assignment Completion on Academic Performance**

Research Question: To what extent do attendance levels and assignment completion rates influence final academic grades?

```python
df['attendance_group'] = pd.cut(df['attendance_percentage'],
                                bins=[0, 70, 90, 100],
                                labels=['Low', 'Medium', 'High'],
                                right=False)
df['assignment_group'] = pd.cut(df['assignments_completed'],
                                bins=[0, 10, 15, df['assignments_completed'].max() + 1],
                                labels=['Low', 'Medium', 'High'],
                                right=False)

pivot = df.pivot_table(values='final_grade',
                       index='attendance_group',
                       columns='assignment_group',
                       aggfunc='mean')

sns.heatmap(pivot, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Final Grade Heatmap (Attendance vs Assignments)")
plt.show()
```

**Insight:**
Students who attend more and complete assignments tend to score slightly higher, but most students already have very high grades.

**Managerial Implication:**
Make grading more challenging and focus more on student engagement to better see the impact of discipline.

**5. Impact of Breaks on Student Productivity**

Research Question:
Does taking more breaks improve productivity, or lead to diminishing returns?

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

# Average productivity per break count
sns.lineplot(
    x='breaks_per_day',
    y='productivity_score',
    data=df,
    marker='o',
    estimator='mean'
)

plt.title("Effect of Breaks on Productivity", fontsize=14, fontweight='bold')
plt.xlabel("Number of Breaks per Day")
plt.ylabel("Average Productivity Score")

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Insight:**
Productivity is highest at around 5 breaks per day, indicating an optimal balance, while too few or too many breaks reduce productivity.

**Managerial Implication:**
* Students should be encouraged to take moderate
breaks (~5 per day)

* Too many breaks may cause distraction, while too few may lead to fatigue
